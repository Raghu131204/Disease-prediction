<h2 style=font-weight:bold;>2.Count plot</h2>

```python
feat = dftrain.columns[:10]  # Select 10 columns

# Create a figure with subplots
fig, axes = plt.subplots(2, 5, figsize=(20, 10))  # 2 rows, 5 columns

# Iterate through features and plot
for i, feature in enumerate(feat):
    row, col = divmod(i, 5)  # Determine row and column for each subplot
    sns.countplot(x=dftrain[feature], ax=axes[row, col])
    axes[row, col].set_title(feature)
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
feat = dftrain.columns[:10]  # Select 10 columns

# Create a figure with subplots
fig, axes = plt.subplots(2, 5, figsize=(20, 10))  # 2 rows, 5 columns

# Iterate through features and plot
for i, feature in enumerate(feat):
    row, col = divmod(i, 5)  # Determine row and column for each subplot
    sns.kdeplot(dftrain[feature], ax=axes[row, col], shade=True)
    axes[row, col].set_title(feature)
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('Density')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
for feature in feat:
    sns.countplot(x=dftrain[feature])
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

```python
def Model(x_train,y_train,x_test,y_test):
    rf.fit(x_train,y_train)
    y_pred=rf.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    print("Accuracy Score:",acc)
    print("Percentage:",acc*100)
```

```python
##the test data used  here is from training data set itself just for validation purpose
Model(X_train,y_train,X_test_vali,y_test_vali)
```

```python
#Here the test data taken from actual text data set
Model(X_train,y_train,X_test_org,y_test_org)
```

```python
##testing the model for individual samples in the data set
row=X_test_org.iloc[1].values.reshape(1,-1)
y_pred=rf.predict(row)
y_pred
```

```python
#confusion matrix for validating data
y_test_pred=rf.predict(X_test_vali)
cm1=confusion_matrix(y_test_vali,y_test_pred)
#print("confusion matrix:\n",cm1)

print("classification report:\n",classification_report(y_test_vali,y_test_pred))
```

```python
def Graph(cm):
    plt.figure(figsize=(10,7))
    sns.heatmap(cm,annot=True)
    plt.xlabel('Predicted value',color='red')
    plt.ylabel('Actual value',color='red')
    plt.title('Confusion Matrix')
    plt.show()
```
